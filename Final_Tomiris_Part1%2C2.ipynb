```python
import matplotlib.pyplot as plt
import matplotlib.image as img
%matplotlib inline
imhead = img.imread('./head.png')
imwheels = img.imread('./wheels.png')
imfigure = img.imread('./figure.png')
imarm  = img.imread('./arm.png')
fig = plt.figure(figsize=(15,10))
plt.subplot(221)
plt.imshow(imarm,cmap='gray')
plt.subplot(222)
plt.imshow(imfigure,cmap='gray')
plt.subplot(223)
plt.imshow(imwheels,cmap='gray')
plt.subplot(224)
plt.imshow(imhead,cmap='gray')
```

```python
from scipy import ndimage as nimg
import numpy as np
#1 arm
gflt = nimg.gaussian_filter(imarm, sigma=20)
fig = plt.figure(figsize=(15,10))
sflt = imarm-gflt
r = np.log(imarm)-np.log(gflt)
r = (r-np.amin(r))/(np.amax(r)-np.amin(r))

flt = ndimage.median_filter(sflt, size=10, mode='reflect', cval=0.0, origin=0)

#2 normal figure
gflt1 = nimg.gaussian_filter(imfigure, sigma=20)
fig1 = plt.figure(figsize=(15,10))
sflt1 = imfigure-gflt1
r1 = np.log(imfigure)-np.log(gflt1)
r1 = (r1-np.amin(r1))/(np.amax(r1)-np.amin(r1))

flt1 = ndimage.median_filter(sflt1, size=10, mode='reflect', cval=0.0, origin=0)


plt.subplot(231)
plt.imshow(imarm,cmap='gray')
plt.subplot(232)
plt.imshow(sflt,cmap='gray')
plt.subplot(233)
plt.imshow(flt,cmap='gray')

plt.subplot(234)
plt.imshow(imfigure,cmap='gray')
plt.subplot(235)
plt.imshow(sflt1,cmap='gray')
plt.subplot(236)
plt.imshow(flt1,cmap='gray')
```
