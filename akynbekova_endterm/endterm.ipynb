```python
from mnist import MNIST
import random
mndata = MNIST('./data')
trimages, trlabels = mndata.load_training()
teimages, telabels = mndata.load_testing()
index = random.randrange(0, len(trimages))  # choose an index ;-)
print('The amount of train images',len(trimages))
print('The amount of test images',len(trimages))
print('The label of random image',trlabels[index],'The random image is',mndata.display(trimages[index]))
print('Images are binary with 28*28 = ',len(trimages[index]))

from keras.models import Sequential
from keras.layers import Dense,Activation,Reshape,Conv2D,Flatten
from keras.layers.core import Dense, Activation, Flatten
from keras.layers.convolutional import Convolution2D, MaxPooling2D
from keras.optimizers import SGD

import os
import gzip
import numpy as np

x_tr = 0
y_tr = 0
x_te = 0
y_te = 0

reshape = (-1, 1, 28, 28)

def to_one_hot(labels, num_classes=10):
    return np.eye(num_classes)[labels]

with gzip.open('./data/train-images-idx3-ubyte.gz', 'rb') as f:
    x_tr = np.frombuffer(f.read(), np.uint8, offset=16)
    x_tr = x_tr.reshape(-1, 28 * 28) / np.float32(256)

with gzip.open('./data/t10k-images-idx3-ubyte.gz') as f:
    x_te = np.frombuffer(f.read(), np.uint8, offset=16)
    x_te = x_te.reshape(-1, 28 * 28) / np.float32(256)

with gzip.open('./data/train-labels-idx1-ubyte.gz', 'rb') as f:
    y_tr = np.frombuffer(f.read(), np.uint8, offset=8)

with gzip.open('./data/t10k-labels-idx1-ubyte.gz', 'rb') as f:
    y_te = np.frombuffer(f.read(), np.uint8, offset=8)


x_tr = x_tr[:-10000]
y_tr = y_tr[:-10000]

y_tr, y_te = [to_one_hot(y) for y in (y_tr, y_te)]

x_tr, x_te = [x.reshape(*reshape) for x in (x_tr, x_te)]

model = Sequential()

model.add(Conv2D(8, (5, 5), activation="sigmoid", input_shape=x_tr.shape[1:], data_format='channels_first'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(100, activation='sigmoid'))
model.add(Dense(10, activation='softmax'))

model.compile(SGD(0.2), 'categorical_crossentropy', metrics=['accuracy'])
model.fit(x_tr, y_tr, batch_size = 10, epochs = 5, verbose = 1, validation_data = (x_te, y_te))

evaluation = model.evaluate(x_te, y_te)

print(evaluation)
```
