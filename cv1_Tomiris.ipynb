```python
#3d task
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
from PIL import Image
im = pltimg.imread('6_26.jpg')
cropped_cat =  im[240:600,  890:1200,:]
plt.imsave('cropped_cat', cropped_cat)
img = Image.open('6_26.jpg', 'r')
img2 = Image.open('cropped_cat', 'r')
img.paste(img2)
img.show()
```

```python
#2nd task
import matplotlib.pyplot as plt
import matplotlib.image as pltimage
from PIL import Image

img = Image.open('6_26.jpg')
area = (800,250, 1250,600)
cropped_img = img.crop(area)

plt.imshow(cropped_img);
plt.show();
```

```python
#1st task
import matplotlib.pyplot as plt
import matplotlib.image as pltimage
from PIL import Image

img = Image.open('6_26.jpg')
r, g, b = img.split()
 
im1 = Image.merge("RGB", (r, g, b))
im2 = Image.merge("RGB", (r, b, g))
im3 = Image.merge("RGB", (g, r, b))
im4 = Image.merge("RGB", (b, g, r))
fig = plt.figure(figsize=(15,10))
plt.subplot(221);
plt.imshow(im1);
plt.subplot(222);
plt.imshow(im2);
plt.subplot(223);
plt.imshow(im3);
plt.subplot(224);
plt.imshow(im4);
plt.show();
```
