```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import matplotlib.image as im
import os
from skimage import io
from scipy import ndimage
img = io.imread(os.path.join("DC.png"), as_grey=True)
filt1 = 1/16*np.matrix([[1, 2, 1],[2, 4, 2],[1, 2, 1]])
filt2 = 1/5*np.matrix([[0, 1, 0],[1, 1, 1],[0, 1, 0]])
filt4 = np.matrix([[-1, -2, -1],[-2, 12, -2],[-1, -2, -1]])
filt5 = np.matrix([[-1, -2, -1],[0, 0, 0],[1, 2, 1]])
```

```python

s__p = 0.5
amount = 0.004
out = np.copy(img)

st = np.ceil(amount * img.size * s__p)
coords = [np.random.randint(0, i - 1, int(st))
        for i in img.shape]
out[coords] = 1
pp = np.ceil(amount* img.size * (1. - s__p))
coords = [np.random.randint(0, i - 1, int(pp))
         for i in img.shape]
out[coords] = 0

flt1 = ndimage.convolve(out, filt1, mode='reflect', cval=0.0)
flt2 = ndimage.convolve(out, filt2, mode='reflect', cval=0.0)
flt3 = ndimage.median_filter(img, 3)
flt4 = ndimage.convolve(out, filt4, mode='reflect', cval=0.0)
flt5 = ndimage.convolve(out, filt5, mode='reflect', cval=0.0)

fig = plt.figure(figsize=(15,10))
plt.subplot(321)
plt.imshow(out, cmap='gray')
plt.subplot(322)
plt.imshow(flt1, cmap='gray')
plt.subplot(323)
plt.imshow(flt2, cmap='gray')
plt.subplot(324)
plt.imshow(flt3, cmap='gray')
plt.subplot(325)
plt.imshow(flt4, cmap='gray')
plt.subplot(326)
plt.imshow(flt5, cmap='gray')
plt.show()
```
