```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import matplotlib.image as im
import os
from skimage import io
from scipy import ndimage

img = io.imread(os.path.join("DC.png"), as_grey=True)
s__p = 0.5
amount = 0.004
out = np.copy(img)
st = np.ceil(amount * img.size * s__p)
coords = [np.random.randint(0, i - 1, int(st))
        for i in img.shape]
out[coords] = 1
pp = np.ceil(amount* img.size * (1. - s__p))
coords = [np.random.randint(0, i - 1, int(pp))
         for i in img.shape]
out[coords] = 0

flt = ndimage.median_filter(img, 4)

fig = plt.figure(figsize=(15,10))
plt.subplot(221)
plt.imshow(flt, cmap='gray')
plt.subplot(222)
plt.imshow((img-out), cmap='gray')
plt.subplot(223)
plt.imshow((img-flt), cmap='gray')
plt.subplot(224)
plt.imshow(img)
plt.show()
```
